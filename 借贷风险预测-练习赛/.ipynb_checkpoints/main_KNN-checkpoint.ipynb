# 数据来源
数据主要包括银行用户信息的脱敏信息。 

数据分为训练数据和测试数据，分别保存在train.csv和test_noLabel.csv两个文件中。 

字段说明如下： 

（1）uid：编号 

（2）x0-x19：银行用户的具体信息，已脱敏处理 

（3）y：银行用户的信用好坏，1表示信用好，0表示信用差。 

如遇数据下载打开乱码问题： 不要用excel打开,用notepad++或者vs code。
文件格式是通用的编码方式utf-8。


网络地址：https://www.dcjingsai.com/v2/cmptDetail.html?id=355

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_noLabel_data=pd.read_csv('test_noLabel.csv')
train_data=pd.read_csv('train.csv')
```

```python
train_data_x = train_data.drop(columns=['y'])
train_data_y = train_data['y']
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
clf = KNeighborsClassifier(n_neighbors=K)
clf.fit(train_data_x, train_data_y)  

result = clf.predict(test_noLabel_data)  # 把X_test放入已经训练好的KNN 模型中。预测X_test被判断出来的标签

```

```python
data = {'uid':test_noLabel_data['uid'],
        'y':[]}
# data = pd.DataFrame(data,index=['one','two','three','four','five'],
#                columns=['姓名','性别','年龄','职业'])
for i in result:
    data['y'].append(i)

show_data =pd.DataFrame(data)
show_data.head(5)
```

```python
# 数据分类情况统计
show_data['y'].value_counts()
```

```python
# 导出数据
show_data.to_csv('predict_KNN.csv',columns=['uid','y'],header=True,index=False)
```
