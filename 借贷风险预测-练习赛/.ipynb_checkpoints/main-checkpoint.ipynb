# 数据来源
数据主要包括银行用户信息的脱敏信息。 

数据分为训练数据和测试数据，分别保存在train.csv和test_noLabel.csv两个文件中。 

字段说明如下： 

（1）uid：编号 

（2）x0-x19：银行用户的具体信息，已脱敏处理 

（3）y：银行用户的信用好坏，1表示信用好，0表示信用差。 

如遇数据下载打开乱码问题： 不要用excel打开,用notepad++或者vs code。
文件格式是通用的编码方式utf-8。

网络地址：https://www.dcjingsai.com/v2/cmptDetail.html?id=355

```python
# train_data_x,train_data_y 分别为数据的特征值和标签值
train_data_x = train_data.drop(columns=['y'])
train_data_y = train_data['y']
```

## 需被验证的数据集

```python
test_noLabel_data.head(30)
```

# 逻辑回归方法

```python

# test_noLabe_data为验证数据的特征值

from sklearn.linear_model import LogisticRegression
# Logistic回归

model = LogisticRegression()
# 创建一个逻辑回归分类器实例并拟合数据。
model.fit(train_data_x, train_data_y )
print("training accuracy :", model.score(train_data_x, train_data_y ))
result = model.predict(test_noLabel_data)
result
```

```python
data = {'uid':test_noLabel_data['uid'],
        'y':[]}
# data = pd.DataFrame(data,index=['one','two','three','four','five'],
#                columns=['姓名','性别','年龄','职业'])
for i in result:
    data['y'].append(i)

show_data =pd.DataFrame(data)
show_data.head(5)
```

```python
# 导出数据
show_data.to_csv('predict_result.csv',columns=['uid','y'],header=True,index=False)
```
